```python
 import numpy as np
import sklearn.datasets as datasets
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
df = datasets.load_iris()
```

# preprocessing

```python
x = df.data
y = df.target
```

# Partitioning the data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0) 
```

# Decision Tree

```python
from sklearn.tree import   DecisionTreeClassifier, plot_tree,export_graphviz
from sklearn.metrics import accuracy_score
```

```python
ct=DecisionTreeClassifier(random_state=0)
```

```python
ct.fit(x_train,y_train);
```

# finding predictions

```python
y_pred=ct.predict(x_test)
```

# view our predictions and actual values together

```python
pd.DataFrame({'orgi':y_test,'predicted':y_pred})
```

# checking the accuracy of our classification

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred)) 
```

```python
print('Accuracy Score =', accuracy_score(y_true=y_test, y_pred=y_pred))
```
